```python
# import library pandas
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder

# load data csv menggunakan pandas
data = 'wbc.csv'
df = pd.read_csv(data)

# menampilkan lima data teratas 
df.head()
```

##### 1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
df = df.drop(["id", "Unnamed: 32"], axis=1, inplace=False)
df.head()

# Pada praktikum ini saya membuang (drop) kolom 'id' dan 'Unnamed: 32', 
# karena menurut saya hanya kedua kolom tersebut yang tidak berpengaruh dalam penentuan kanker tersebut termasuk jinak atau ganas
```

#### 2. Lakukan proses encoding pada kolom "diagnosis".

```python
# saya menggunakan label encoding karena tidak memperhatikan urutan
le = LabelEncoder() 
df['diagnosis'] = le.fit_transform(df['diagnosis'])
df.head()
```

#### 3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
std = StandardScaler()

# saya lakukan looping sebanyak jumlah kolom 
# untuk kolom diagnosis, karena itu merupakan hasil encoding maka tidak saya lakukan standarisasi 
for column in df.columns :
    if(column != 'diagnosis'):
        df[column] = std.fit_transform(df[[column]])

df.head()
```

#### 4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
from sklearn.model_selection import train_test_split
# Split data training dan data testing
# Rasio yang akan kita gunakan adalah 8:2
df_train, df_test = train_test_split(df, test_size=0.2, random_state=0, stratify=df['diagnosis'])

# Cek masing-masing ukuran data
print(f'Jumlah label data asli:\n{df.diagnosis.value_counts()}')
print(f'Jumlah label data train:\n{df_train.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{df_test.diagnosis.value_counts()}')
```
